## Polynomial Regression Model with degree One 

```python
import numpy as np
import matplotlib.pyplot as plt

#data
X_data= np.array([[1,12],
            [1,23],
            [1,34],
            [1,45], 
            [1,56], 
            [1,67], 
            [1,78], 
            [1,89], 
            [1,123], 
            [1,134]])
Y_data=np.array([240, 1135, 2568, 4521, 7865, 9236, 11932, 14589, 19856, 23145])



#Calculating the result matrix
transpose_matrix=np.transpose(X_data)
print("\nTranspose_matrix:\n",transpose_matrix)

multi=np.matmul(transpose_matrix,X_data)
print("\nMultiplied_Result:\n",multi)

inverse_matrix = np.linalg.inv(multi)
print("\ninverse_matrix:\n",inverse_matrix)

new=np.matmul(transpose_matrix,Y_data)

result=np.matmul(inverse_matrix,new)
print("\n\nRESULT:",result)
print(len(X_data))

 
```

```python
def prediction(X,R,Y):
    y = np.zeros_like(Y)
    for i in range(len(X_data)): 
        y[i] = R[1]*X[i][1]+R[0] #Taking the coefficients from the result vector
    return y

#Predicted value
y_predict = prediction(X_data,result,Y_data)
print(y_predict)

def SSE(x, Y,r,y_predict):
    cost = np.sum((y_predict - Y) ** 2) #y_predict adn Y are same dimension vectors so easy to do calculation
    return cost

#Sum square error SSE
sum_sqr_error=SSE(X_data, Y_data,result,y_predict)
print("SSE:",sum_sqr_error)

#Y mean and SST
def SST(Y_data):
    y_mean=np.mean(Y_data)
    k=(Y_data-y_mean)**2
    sst=np.sum(k)
    return sst

sst=SST(Y_data)
print("SST:",sst)

#R_square value
R_square=1-(sum_sqr_error/sst)
print("R_Square:",R_square)
```

```python
#plot 
x=X_data[:,1]
print("x:",x)
plt.plot(x,y_predict,label="x_data vs Prediction_data")
plt.scatter(x,Y_data,color='red',label="X_data vs Observation_data")
plt.title("Degree-1")
plt.xlabel("X_data")
plt.ylabel("Y_data")
plt.legend()
```

## Polynomial Regression using degree three

```python
import numpy as np
import matplotlib.pyplot as plt

#data
X_initial= np.array([[1,12],
            [1,23],
            [1,34],
            [1,45], 
            [1,56], 
            [1,67], 
            [1,78], 
            [1,89], 
            [1,123], 
            [1,134]])
Y_data=np.array([240, 1135, 2568, 4521, 7865, 9236, 11932, 14589, 19856, 23145])

temp=X_initial[:,1]**3
print(f"Dimention:{temp.ndim}\ntemp:\n{temp}")

temp=np.expand_dims(temp,axis=1)#change dimention from 1 to 2 for append to X_Data
print(f"Dimention:{temp.ndim}\ntemp:\n{temp}")

X_data=np.append(X_initial,temp,axis=1)
print(f"x_Data:\n",X_data)

#Calculating the result matrix
transpose_matrix=np.transpose(X_data)
print("\nTranspose_matrix:\n",transpose_matrix)

multi=np.matmul(transpose_matrix,X_data)
print("\nMultiplied_Result:\n",multi)

inverse_matrix = np.linalg.inv(multi)
print("\ninverse_matrix:\n",inverse_matrix)

new=np.matmul(transpose_matrix,Y_data)
coefficient=np.matmul(inverse_matrix,new)
print("\n\coefficient:",coefficient)



```

```python
# Prediction or y_hat function
def prediction(X,R,Y):
    y = np.zeros_like(Y)
    for i in range(len(X_data)): 
        y[i] = R[2]*X[i][2]+R[1]*X[i][1]+R[0]
    return y

#Predicted array y_hat 
y_predict = prediction(X_data,coefficient,Y_data)
print("y_predict:",y_predict)

#Sum_square_error calculation
def SSE(x, Y,r,y_predict):
    cost = np.sum((y_predict - Y) ** 2)
    return cost

sum_sqr_error=SSE(X_data, Y_data,coefficient,y_predict)
print("SSE:",sum_sqr_error)

#SST calculation
def SST(Y_data):
    y_mean=np.mean(Y_data)
    k=(Y_data-y_mean)**2
    sst=np.sum(k)
    return sst

sst=SST(Y_data)
print("SST:",sst)

#R_Square
R_square=1-(sum_sqr_error/sst)
print("R_square:",R_square)

```

```python
#plot 
x=X_data[:,1]
print("x:",x)
plt.plot(x,y_predict,label="x_data vs Prediction_data")
plt.scatter(x,Y_data,color='red',label="X_data vs Observation_data")
plt.title("Degree-3")
plt.xlabel("X_data")
plt.ylabel("Y_data")
plt.legend()
```
