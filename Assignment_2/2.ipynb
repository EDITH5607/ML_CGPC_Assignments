```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
df = pd.read_csv('diabetes.csv')
df.head(5)
```

```python
X = df.drop('Outcome',axis = 1)
X
```

```python
y = df['Outcome']
y, df['Outcome'].value_counts(normalize=True)*100
```

```python
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'SVM': SVC(kernel='linear', C=1000, random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}
splits = {
    '80/20': 0.8,
    '70/30': 0.7,
    '60/40': 0.6
}
```

```python
model_preds ={}
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size =  split, random_state = 10, stratify=y )
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    model_preds[model_name] = y_predict
    
```
